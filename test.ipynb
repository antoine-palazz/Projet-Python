```python
def format_nom_url(nom):
    mots = nom.split()
    if len(mots) == 1:
        # Cas où il n'y a qu'un seul mot dans le nom
        nom_formatte = mots[0].lower()
    else:
        # Cas où il y a plusieurs mots dans le nom
        nom_famille = mots[-1].lower()
        prenom_nom = '-'.join(mots[:-1]).lower()
        nom_formatte = f"{prenom_nom}_{nom_famille}"
    return nom_formatte
```
